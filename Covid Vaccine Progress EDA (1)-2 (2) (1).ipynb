## Deal With Null Values

```python
## Get Null Percentage
for col in df.columns:
    missing = np.mean(df[col].isnull()) * 100
    print(f'{col} - {round(missing,2)}%')
```

```python
# Delete All Null Values 
df.dropna(subset=['people_vaccinated'], how='all', inplace=True)
```

```python
## Get Null Percentage
for col in df.columns:
    missing = np.mean(df[col].isnull()) * 100
    print(f'{col} - {round(missing,2)}%')
```

```python
# the new data shape after remove null 
df.shape
```

```python
# Delete All Null Columns and Unnecassry Columns
df.drop('source_website', axis=1, inplace=True)
```

### Visualize The Data

```python
d = df.groupby(by=['country'], as_index=False)['people_fully_vaccinated'].sum()
d.sort_values(by='people_fully_vaccinated', ascending=False,inplace=True)
d.head(10)
```

```python
# america is the most country showed an interest in vaccine then india 
d = df.groupby(by=['country'], as_index=False)['people_fully_vaccinated'].sum()
d.sort_values(by='people_fully_vaccinated', ascending=False,inplace=True)
d.head(10).plot(kind="bar")
```

```python
# most of the data comes from government of jersey then public health agency of sweeden 
df['source_name'].value_counts().sort_values().head(10).plot.barh()
```

```python
# most of the people are getting vaccine 
plt.scatter(df['total_vaccinations'],df['people_fully_vaccinated'])
plt.xlabel('Total Vaccination ')
plt.ylabel('People Fully Vaccinated')
```
